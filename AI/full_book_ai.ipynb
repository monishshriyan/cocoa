## Practical 3
Aim: Implement A* search algorithm for Romanian map problem.


```python
import queue as Q
dict_hn = { "Arad": 336, "Bucharest": 0, "Craiova": 160, "Dobreta": 242, "Eforie": 161,"Fagaras": 178, "Giurgiu": 77, "Hirsova": 151, "Iasi": 226, "Lugoj": 244,"Mehadia": 241, "Neamt": 234, "Oradea": 380, "Pitesti": 98, "Rimnicu_Vilcea": 193,"Sibiu": 253, "Timisoara": 329, "Urziceni": 80, "Vaslui": 199, "Zerind": 374}
dict_gn = {
    "Arad": {"Zerind": 75, "Timisoara": 118, "Sibiu": 140},
    "Bucharest": {"Urziceni": 85, "Giurgiu": 90, "Pitesti": 101, "Fagaras": 211},
    "Craiova": {"Dobreta": 120, "Rimnicu_Vilcea": 146, "Pitesti": 138},
    "Dobreta": {"Craiova": 120, "Mehadia": 75},
    "Eforie": {"Hirsova": 86},
    "Fagaras": {"Sibiu": 99, "Bucharest": 211},
    "Giurgiu": {"Bucharest": 90},
    "Hirsova": {"Eforie": 86, "Urziceni": 98},
    "Iasi": {"Neamt": 87, "Vaslui": 92},
    "Lugoj": {"Mehadia": 70, "Timisoara": 111},
    "Mehadia": {"Dobreta": 75, "Lugoj": 70},
    "Neamt": {"Iasi": 87},
    "Oradea": {"Zerind": 71, "Sibiu": 151},
    "Pitesti": {"Rimnicu_Vilcea": 97, "Bucharest": 101, "Craiova": 138},
    "Rimnicu_Vilcea": {"Pitesti": 97, "Craiova": 146, "Sibiu": 80},
    "Sibiu": {"Oradea": 151, "Fagaras": 99, "Rimnicu_Vilcea": 80},
    "Timisoara": {"Arad": 118, "Lugoj": 111},
    "Urziceni": {"Bucharest": 85, "Vaslui": 142, "Hirsova": 98},
    "Vaslui": {"Iasi": 92, "Urziceni": 142},
    "Zerind": {"Arad": 75, "Oradea": 71}
}
def get_fn(citystr):
    """ Calculating f(n) = g(n) + h(n) for a given path """
    cities = citystr.split(" , ")
    gn = 0
    hn = dict_hn[cities[-1]]
    for i in range(len(cities) - 1):
        gn += dict_gn[cities[i]][cities[i + 1]]
    return gn + hn
def expand(cityq):
    global result
    tot, citystr, thiscity = cityq.get()
    if thiscity == goal:
        result = citystr + " : : " + str(tot)
        return
    for cty in dict_gn[thiscity]:
        cityq.put((get_fn(citystr + " , " + cty), citystr + " , " + cty, cty))
    expand(cityq)
def main():
    cityq = Q.PriorityQueue()
    cityq.put((get_fn(start), start, start))
    expand(cityq)
    print("The A* path with the total is:")
    print(result)
if __name__ == "__main__":
    start = 'Arad'
    goal = 'Bucharest'
    result = ''
    main()
```

## Practical 7
Aim:  Python program to demonstrate how to compute the output of a neuron with different activation function.
 

```python
import numpy as np
def neuron_output(inputs,weights,bias,activation_function):
    weighted_sum=np.dot(inputs,weights)+bias
    print("Weighted sum is :",weighted_sum)
    #Apply the specified activation function
    if activation_function =='sigmoid':
        output= 1/(1 + np.exp(-weighted_sum))
    elif activation_function =='relu':
        output = max(0,weighted_sum)
    elif activation_function =='tanh':
        output = np.tanh(weighted_sum)
    elif activation_function =='linear':
        output = weighted_sum
    else:
        raise ValueError("Unknown activation function")
    print("Final Ouptut: ", output)
    return output

inputs=[0.3,0.4]
weights=[0.2,0.6]
bias= 0.5


#Compute the neuron output with different activation functions
print("Activation Function: Sigmoid")
output_sigmoid=neuron_output(inputs,weights,bias,activation_function='sigmoid')

print("\nActivation Function: ReLU")
output_relu=neuron_output(inputs,weights,bias,activation_function='relu')

print("\nActivation Function: TanH")
output_tanh=neuron_output(inputs,weights,bias,activation_function='tanh')

print("\nActivation Function: Linear")
output_linear=neuron_output(inputs,weights,bias,activation_function='linear')
```

## Practical 9
Aim: Python program to implement Depth Limited Search Algorithm. 
 

```python
def get_neighbors(city):
    graph = {
        'Arad': {'Zerind': 75, 'Timisora': 118, 'Sibiu': 140},
        'Bucharest': {'Urziceni': 85, 'Giurgiu': 90, 'Pitesti': 101, 'Fagaras': 211},
        'Craiova': {'Dobreta': 120, 'Riminca_Vilcea': 146, 'Pitesti': 138},
        'Dobreta': {'Craiova': 120, 'Mehadia': 75},
        'Eforie': {'Hirsova': 86},
        'Fagaras': {'Sibiu': 99, 'Bucharest': 211},
        'Giurgiu': {'Bucharest': 90},
        'Hirsova': {'Eforie': 86, 'Urziceni': 98},
        'Iasi': {'Neamt': 87, 'Vaslui': 92},
        'Lugoj': {'Mehadia': 70, 'Timisora': 111},
        'Mehadia': {'Dobreta': 75, 'Lugoj': 70},
        'Neamt': {'Iasi': 87},
        'Oradea': {'Zerind': 71, 'Sibiu': 151},
        'Pitesti': {'Riminca_Vilcea': 97, 'Bucharest': 101, 'Craiova': 138},
        'Riminca_Vilcea': {'Pitesti': 97, 'Craiova': 146, 'Sibiu': 80},
        'Sibiu': {'Oradea': 151, 'Fagaras': 99, 'Riminca_Vilcea': 80},
        'Timisora': {'Arad': 118, 'Lugoj': 111},
        'Urziceni': {'Bucharest': 85, 'Vaslui': 142, 'Hirsova': 98},
        'Vaslui': {'Iasi': 92, 'Urziceni': 142},
        'Zerind': {'Arad': 75, 'Oradea': 71}
    }
    return graph.get(city, {})

def dls(start, goal, depth, path=[]):
    if start == goal:
        return path + [start]
    if depth <= 0:
        return []
    for neighbor, _ in get_neighbors(start).items():
        new_path = path + [start]
        result = dls(neighbor, goal, depth - 1, new_path)
        if result:
            return result
    return []

def depth_limited_search(start, goal, max_depth):
    return dls(start, goal, max_depth)

# Example usage
start_city = 'Arad'
goal_city = 'Bucharest'
max_depth = int(input("Enter Depth"))

path = depth_limited_search(start_city, goal_city, max_depth)
if path:
    print(f"Path found from {start_city} to {goal_city} within depth {max_depth}:")
    print(" -> ".join(path))
else:
    print(f"No Path exists from {start_city} to {goal_city} within depth {max_depth}.")
```
